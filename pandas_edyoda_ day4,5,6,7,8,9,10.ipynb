```python
# cut and qcut
import pandas as pd
import numpy as np
df = pd.DataFrame({'Age':np.random.randint(10, 90, 20)})
```

```python
df
```

```python
# cut()/qcut() are used to convert numerical values to categorical values. Also known as binning/bucketing
labels = ['underage', 'adult', 'mature', 'sr. citizen'] 
bin = [0, 18, 30, 60, 100]
```

```python
df['age_group'] = pd.cut(df['Age'], bins = bin, labels = labels)
```

```python
df.sort_values('Age')
```

```python
df['qage_group'] = pd.qcut(df['Age'], q=4, labels = labels)
```

```python
df_new = df.sort_values(['Age'])
```

```python
df_new
```

```python
# Finding duplicate entries exist or not
df_new.duplicated(['age_group', 'qage_group'])
```

```python
df_new.drop_duplicates(['age_group'])
```

```python
df_new.drop_duplicates(['age_group'], keep = False)
```
