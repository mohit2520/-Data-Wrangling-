```python
def extract_number(x):
  # x = np.any([int(i[:3])>30 for i in x.split(',')[1:]])
  x = x.split(',')
  l = []
  for i in x:
    try:
      if i[:3].isdigit():
        l.append(i[:3])
    except :
      pass
  return np.any([int(i)>30 for i in l])
```
